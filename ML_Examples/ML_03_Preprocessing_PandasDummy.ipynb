```python
db.columns=["yil","kasa","mil","motor","beygir","vites","yakit","sahip","fiyat","ppy"]
```

```python
db.head()
```

```python
db['motor']=db['motor'].str.replace("L","")
```

```python
db['motor']=pd.to_numeric(db['motor'])
```

```python
db.head()
```

```python
db.info()
```

```python
db=pd.get_dummies(db,columns=['kasa','vites','yakit'],drop_first=True)
```

```python
db.head()
```

```python
db.info()
```

```python
y=db[['fiyat']]
x=db.drop("fiyat",axis=1)
```

```python
lm=LinearRegression()
model=lm.fit(x,y)
```
